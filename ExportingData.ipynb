**Exporting Data**

We can export documents in a variety of formats using the suitcase package

```python

import databroker
# Connect to the sissy2 catalog
db = databroker.catalog["sissy2"]
```

```python
run = db["3f8ce47"]
```

```python
run.primary.read()
```

```python
docs=run.documents(fill='yes')
import suitcase.csv
suitcase.csv.export(docs, '')
```

This will save the data to the local directory. You can then download it onto your machine (right click on the file and press download)
